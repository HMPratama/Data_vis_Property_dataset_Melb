### **Nomor 7**

Tampilkan visualisasi yang menggambarkan perubahan jumlah rumah yang dijual dari Jan 2016 - Des 
2017!

```python
df
```

```python
df['MonthYear'] = df['Date'].apply(lambda x: str(x)[-7:])
df['MonthYear'].head()
```

```python
df[['MonthYear']].sort_values(by='MonthYear',ascending=True)
```

#### Note: Data yang terdeteksi hanya sampai Dec 2017

```python
df_date = df.groupby(by='MonthYear')[['Regionname']].count().reset_index()
df_date
```

```python
plt.figure(figsize=(10,5))
plt.title('Trend of Trading House by Month')
plt.ylabel('amount sold of House')
plt.xlabel('Period')
sns.lineplot(data = df_date,
             x = 'MonthYear',
             y = 'Regionname',
             errorbar=('ci',0),  # menghilangkan confidence interval
             marker='d',
             markersize=15,
             color= 'blue',
             estimator='count') # estimator by default : mean

# Mengatur posisi Xticks
plt.xticks(ticks=df_date['MonthYear'], 
           labels=[dt.strftime('%b-%Y') for dt in df_date['MonthYear']],
           rotation= 90)

for x, y in zip(df_date['MonthYear'],df_date['Regionname']):
    plt.text(s=f'{y:^.0f}', x=x, y=y-80)

plt.show()
```

**Insight:**
- trend penjualan rumah dari Jan-2016 sampai Apr-2016 naik sangat signifikan
- trend penjualan Apr-2016 sampai sep-2017 sangat fluktuatif

##### Harga untuk "t" (Townhouse)

```python
# Q3 = Sebagai "t"(TownHouse) outlier
df['Price'].describe()['75%']
```

```python
q1 = df['Price'].describe()['25%']
q3 = df['Price'].describe()['75%']
iqr = q3 - q1
iqr

lowerbound = q1 - (1.5*iqr)
upperbound = q3 + (1.5*iqr)

outliers_count=df[(df['Price']< lowerbound) | (df['Price']> upperbound)]['Price'].count()
print('Number of outliers count:',outliers_count)
```
